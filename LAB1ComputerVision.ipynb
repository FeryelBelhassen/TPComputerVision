### 1- Image Augmentation with OpenCV

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
image = cv2.imread('C:/Users/HP/Documents/Feryel/3DNI/LAB1_CV_3DNI_2023/bird.jpg')
filename = 'bird.jpg'
dim = (200, 100)
rows, cols, ch = image.shape
M = cv2.getAffineTransform(pts1, pts2) 
pts1 = np.float32([[50, 50],
                   [200, 50], 
                   [50, 200]])
  
pts2 = np.float32([[10, 100],
                   [200, 50], 
                   [100, 250]])
Kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

def image_augmentation(image, path):
    
#Flipping:
    image=cv2.flip(image, 0)
    plt.subplot(2,4,1),plt.imshow(image, cmap='gray')
    
#Cropping:
    image_croc = image[50:150, 20:100]
    plt.subplot(2,4,2),plt.imshow(image_croc, cmap='gray')
    
#Scaling:
    resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    plt.subplot(2,4,3),plt.imshow(resized, cmap='gray')
    
#Rotating:
    rotated = cv2.warpAffine(image, M, (cols, rows))
    plt.subplot(2,4,4), plt.imshow(rotated)
    
#Brightness:
    matrix = np.ones(image.shape, dtype = "uint8") * 120 
    image_bri = cv2.add(image, matrix)
    plt.subplot(2,4,5), plt.imshow(image_bri)
    
#Median Blur:    
    median_img=cv2.medianBlur(image,11)
    plt.subplot(2,4,6), plt.imshow(median_img)

#Sharpening:
    sharpened = cv2.filter2D(image, -1, Kernel)
    plt.subplot(2,4,7),plt.imshow(sharpened)
    plt.show()
    
image_augmentation(image,'C:/Users/HP/Documents/Feryel/3DNI/LAB1_CV_3DNI_2023/bird.jpg')
cv2.imwrite(filename,image)







    
```
